```python
X_train = pd.read_csv('X_train.csv')
y_train = pd.read_csv('y_train.csv')
X_test = pd.read_csv('X_test.csv')
y_test = pd.read_csv('y_test.csv')

print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)

frames_X = [X_train, X_test]
frames_y = [y_train, y_test]

X = pd.concat(frames_X)
y = pd.concat(frames_y)

print(X.shape)
print(y.shape)
```

## SMOTE Oversampling

```python
print(Counter(y['OverallPoF']))

sm = ADASYN(random_state=42, sampling_strategy = {0:5000, 1:5000})

X_smote, y_smote = sm.fit_resample(X, y)

print(Counter(y_smote['OverallPoF']))
```
