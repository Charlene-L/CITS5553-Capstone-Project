## Data Preprocessing

```python
data = pd.read_csv('Mall_Customers.csv')
print(data.shape)
data.head()
```

```python
data = data.drop('CustomerID', axis=1)
data = data.rename(columns={
    'Annual Income (k$)':'Income',
    'Spending Score (1-100)':'Spend_Score'
})
df = pd.get_dummies(data) # create indicator variables

print(df.shape)
df.head()
```

## Scale Data

```python
scaler = MinMaxScaler()
df_scaled = scaler.fit_transform(df)
df_scaled = pd.DataFrame(df_scaled, columns = df.columns)

print(df_scaled.shape)
df_scaled.head()
```

## DBSCAN

```python
outlier_detection = DBSCAN(
    eps = 0.2,
    metric = "euclidean",
    min_samples = 5,
    n_jobs = -1
)

clusters = outlier_detection.fit_predict(df_scaled.iloc[:,0:3]) # -1 is outlier, 0 is normal
print(clusters.shape)
clusters
```

```python
cmap = cm.get_cmap('Set1')
df_scaled.plot.scatter(x='Spend_Score', y='Income', c=clusters, cmap=cmap)
```

## DBSCAN on WBA Data

```python
outlier_detection = DBSCAN(
    eps = 1.31,
    metric = "euclidean",
    min_samples = 5,
    n_jobs = -1
)

clusters = outlier_detection.fit_predict(wba_scaled) # -1 is outlier, num indicates subgroup

wba_y_pred = clusters
for i in range(len(wba_y_pred)):
  if wba_y_pred[i] != -1:
    wba_y_pred[i] = 0
for i in range(len(wba_y_pred)):
  if wba_y_pred[i] == -1:
    wba_y_pred[i] = 1

print(confusion_matrix(wba_y_true, wba_y_pred)) # columns are predictions, rows are truth
print(classification_report(wba_y_true, wba_y_pred))
```

## Isolation Forest

```python
rs = np.random.RandomState(0) # psuedo random number generator
clf = IsolationForest(max_samples=100, random_state=rs, contamination=0.1)
clf.fit(df_scaled)
if_scores = clf.decision_function(df_scaled)
if_anomalies = clf.predict(df_scaled)
if_anomalies = pd.Series(if_anomalies).replace([-1,1], [1,0])
if_anomalies = df_scaled[if_anomalies == 1]
```

```python
plt.figure(figsize=(12, 8))
plt.hist(if_scores)
plt.title('Histogram of Avg Anomaly Scores: Lower => More Anomalous')
```

```python
cmap = np.array(['white','red'])
plt.scatter(df_scaled['Income'], df_scaled['Spend_Score'], c='white', s=20, edgecolor='k')
plt.scatter(if_anomalies['Income'], if_anomalies['Spend_Score'], c='red')
plt.xlabel('Income')
plt.ylabel('Spend Score')
plt.title('Isolation Forests - Anomalies')
```
