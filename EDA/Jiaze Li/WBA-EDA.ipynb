## random forest & PCA

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.model_selection as ms
from sklearn.svm import SVR
from sklearn.model_selection  import train_test_split
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from pandas.plotting import scatter_matrix
from sklearn.metrics import mean_squared_error
from math import sqrt
```

```python
attribute.head()
```

```python
data = attribute[['Gender','AgeM','ECV','TPP','SC','TympPoF','OAE1','OAE1.4','OAE2','OAE2.8','OAE4','OAEPoFC','Ear coded']]
target = attribute['OverallPoF']
```

```python
data.head()
```

```python
# Perform a random 85/15 split of the data to form a training set and a test set
X_train,X_test,y_train,y_test = ms.train_test_split(data,target,train_size=0.85, random_state=4)
print(len(X_train), "train +", len(X_test), "test")
print("X_train_shape:", X_train.shape, " y_train_shape:", y_train.shape)
print("X_test_shape:", X_test.shape,"  y_test_shape:", y_test.shape)
```

```python
from sklearn.model_selection  import train_test_split, GridSearchCV
from sklearn.metrics import mean_squared_error
from sklearn.ensemble import RandomForestRegressor
def regressor(regressor_name,param):
    # Parameters are tuned with GridSearchCV.
    clf = GridSearchCV(regressor_name(n_estimators=500,max_samples=0.7, bootstrap=True),param_grid=param,cv=3)
    clf.fit(X_train, y_train)#Train the regressor with the training set
    return clf.best_params_,clf.best_score_ #To get the best result of parameter
```

```python
rf = RandomForestRegressor(n_estimators=500,max_samples=0.7, bootstrap=True,min_samples_leaf=2,max_features=5,random_state=42)
rf.fit(X_train, y_train)
score = 100 * rf.score(X_train, y_train)
print('Accuracy:', round(score, 2), '%.')
```

```python
y_train_true,y_train_pred=y_train,rf.predict(X_train)
```

```python
y_train_pred
```

```python
y_train_pred
```

```python
# Calculate the Root Mean Square Error(RMSE)
rmse = sqrt(mean_squared_error(y_train_true,y_train_pred))
# Print out the RMSE
print ("RMSE for train data set:", rmse)
```

```python
# the RMSE of the predictions on the test sets
y_test_true, y_test_pred = y_test, rf.predict(X_test)
# the quality wine values must be integers
y_test_pred = np.round(y_test_pred)
# Calculate the Root Mean Square Error(RMSE)
rmse_test = sqrt(mean_squared_error(y_test_true, y_test_pred))
# Print out the RMSE
print("RMSE for test data set:", rmse_test)
```

```python
# feature importance
importances = rf.feature_importances_
feat_labels = ['Gender','AgeM','ECV','TPP','SC','TympPoF','OAE1','OAE1.4','OAE2','OAE2.8','OAE4','OAEPoFC','Ear coded']
indices = np.argsort(importances)[::-1]
for f in range(X_train.shape[1]):
    print("%2d) %-*s %f" % (f + 1, 30, feat_labels[indices[f]], importances[indices[f]]))
```

Only those features whose importance is higher than 5% are retained 

Finally PCA and Random Forest algorithms were used to classify the dataset, look at the relative importance of the model input features, perform data dimensionality reduction and the results showed that TympType , TPP , OAE were relatively more important.
